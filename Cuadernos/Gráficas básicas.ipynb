Para el gráfico circular o de sectores, primero se debe verificar que la variable a trabajar sea de tipo cualitativo en escala nominal. En el caso de la base ejemplo, se utilizará la variable Continente.

Para seguir este paso a paso puede utilizar el archivo de Excel DATOS GUIA 1 que se encuentra en la carpeta. Importe los datos como se vio en el material anterior.

Se debe recordar que el gráfico circular representa la proporción de unidades
observacionales que pertenecen a cada categoría de respuesta. En este
sentido, los pasos para la construcción son los siguientes:

1. Construcción de una tabla de frecuencias
2. Cálculo de proporciones
3. Construcción del gráfico
![ ](https://raw.githubusercontent.com/jazaineam1/UDLSDE/master/Cuadernos/Imagenes/G2.JPG)

En la ventana de Environment se observa: se creo un objeto tipo tabla de frecuencias TABLA1, un objeto TABLA2 de proporciones.
En la pestaña Plots, se observa el gráfico construido

![ ](https://raw.githubusercontent.com/jazaineam1/UDLSDE/master/Cuadernos/Imagenes/G3.jpg)

Siga las siguientes instrucciones (una a una) y escriba en su cuaderno que acción genera cada una de ellas:

```r
pie(TABLA2, main="Proporcion de paises por contienente")
pie(TABLA2, main="Proporcion de paises por contienente", labels="Africa,_26%", "America_19%", "Asia_20%", "Europa_30%", "Oceania_5%")
pie(TABLA2, main="Proporcion de paises por contienente", labels="Africa,_26%", "America_19%", "Asia_20%", "Europa_30%", "Oceania_5%", col=c("blue", "purple", "red", "orange", "yellow")   
 ```
 Si se pretende comparar como están distribuidos los países por continente para Grupo de ingresos Mediano bajo e ingresos Mediano alto
 
 ```r
par(mfrow=c(1,2))
pie(prop.table(table(Continente['Grupo Ingeso'==2])), main="Ingreso mediano bajo") 
pie(prop.table(table(Continente['Grupo Ingeso'==3])), main="Ingreso mediano alto")
```

La función par indica que se van a comparar varios gráficos, c(f,c) indica cuántas filas y cuántas columnas de gráficos se necesitan.
En este caso se muestra un ejemplo donde las tablas de frecuencia no fueron guardadas de manera previa, sino que se generan dentro de la misma instrucción del gráfico.
Los $ [ ]$ se utilizan si se quiere utilizar información filtrada por la categoría de otra variable.

## Gráfica de Barras

Para dar un ejemplo de construcción de un gráfico de Barras se utilizará la variable Income_Group, ya que cumple con las características de ser de tipo cualitativo y estar sus datos en escala ordinal.
![ ](https://raw.githubusercontent.com/jazaineam1/UDLSDE/master/Cuadernos/Imagenes/G5.jpg)

Se debe recordar que el gráfico de barras puede representar el conteo o la proporción de unidades observacionales que pertenecen a cada categoría de respuesta. En este sentido, los pasos para la construcción son los siguientes:

1. Construcción de una tabla de frecuencias
2. Cálculo de proporciones
3. Construcción del gráfico

![ ](https://raw.githubusercontent.com/jazaineam1/UDLSDE/master/Cuadernos/Imagenes/G6.JPG)
Ingresamos las instrucciones en ventana Console, se generan los objetos TABLA3 y TABLA4 como se puede observar en la ventana Environment. Con la instrucción par(), indicamos que se van a construir dos gráficos de barras.

Se observa que no hay cambio en la forma, si se utiliza el conteo o se utiliza la proporción.


En el gráfico se puede apreciar que el eje horizontal no está construido de manera correcta, pues tenemos una variable en escala ordinal y el eje no lo representa.
![ ](https://raw.githubusercontent.com/jazaineam1/UDLSDE/master/Cuadernos/Imagenes/G7.jpg)
Para solucionar esto se utilizará la columna Grupo ingreso, la cual es una codificación de Income_Group
![ ](https://raw.githubusercontent.com/jazaineam1/UDLSDE/master/Cuadernos/Imagenes/G8.jpg)

```r
barplot(table('Grupo Ingeso'), main="Distribucion de paises", xlab="Grupo de Ingreso", ylab="Cantidad de paises", col="green",names.arg = c("Bajo","Mediano bajo", "Mediano Alto", "Alto"))
```
![ ](https://raw.githubusercontent.com/jazaineam1/UDLSDE/master/Cuadernos/Imagenes/g9.jpg)

## Histograma

Se debe recordar que el histograma representa la tendencia de los datos para una variable de tipo cuantitativa, en el caso de RStudio no es necesario construir una tabla para elaborar el histograma. Como ejemplo se utiliza la variable %Importación de alimentos.

```R
hist('Importaciones de alimentos (% de importaciones de mercaderías)', main="Importacion de alimentos", xlab="Porcentaje de importación", ylab="Cantidad de paises",col="pink")
```
